```
#import the networkx library
import networkx as nx 
  
#importing the matplotlib library for plotting the graph 
import matplotlib.pyplot as plt 
import sys
import collections

#creat a graph with 2 nodes and 1 link  
G0=nx.Graph()
G0.add_node(1)
G0.add_node(2)
G0.add_edge(1,2)

nx.draw(G0, with_labels=True)
print('Two node network example:') 
plt.show() 
```

```
##########################################
#create a random graph  
G1= nx.erdos_renyi_graph(30,0.1)
#plot the graph
nx.draw(G1, with_labels=False) 
print('Erdos Renyi network example 1:') 
plt.show() 

###########################################
#change the nodes and probability for links
G2 = nx.erdos_renyi_graph(100,0.5) 
#draw network
nx.draw(G2, with_labels=True) 
print('Erdos Renyi network example 2:')
plt.show()
```

```
###########################################
nx.connected_components(G1)
print('Shortest path example:')
nx.shortest_path(G2,3,10)
```

```
###########################################
ws=nx.watts_strogatz_graph(60,3,0.5)
#draw network
nx.draw(ws, with_labels=True) 
print('Watts Strogatz network example:') 
plt.show()
```

```
print('Shortest path example:')
nx.shortest_path(ws,34,54)
```

```
###########################################
ba=nx.barabasi_albert_graph(200,3)
nx.draw(ba, with_labels=True) 
print('Barabasi Albert network example:') 
plt.show()
```

```
###########################################
degree_sequence = sorted([d for n, d in ba.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d + 0.4 for d in deg])
ax.set_xticklabels(deg)

# draw graph in inset
plt.axes([0.4, 0.4, 0.5, 0.5])
Gcc = ba.subgraph(sorted(nx.connected_components(ba), key=len, reverse=True)[0])
pos = nx.spring_layout(ba)
plt.axis('off')
nx.draw_networkx_nodes(ba, pos, node_size=20)
nx.draw_networkx_edges(ba, pos, alpha=0.4)

plt.show()
############################################
# Homework: create histograms for various ER and WS networks
```

```

############################################
# Reading and writing graphs
G3 = nx.grid_2d_graph(5, 5)  # 5x5 grid

# print the adjacency list
for line in nx.generate_adjlist(G3):
    print(line)
# write edgelist to grid.edgelist
nx.write_edgelist(G3, path="grid.edgelist", delimiter=":")
# read edgelist from grid.edgelist
G4 = nx.read_edgelist(path="grid.edgelist", delimiter=":", nodetype=str)

nx.draw(G4, with_labels=True)
plt.show()
#############################################
```

```
nx.shortest_path(G4, '(4, 1)', '(0, 0)')
```
